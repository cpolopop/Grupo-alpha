```python
#Importamos la base de datos en formato CSV
import pandas as pd
link="https://github.com/Team-AlphaP/TAREA/raw/main/dataUN%20-%20HDR21-22_Statistical_Annex_HDI_Table.csv"
data=pd.read_csv(link)
```

```python
#Imprimimos los nombres de las columnas
data.columns.to_list()
```

```python
#Para eliminar espacios en los títulos de las columnas
data = data.rename(columns=lambda x: x.replace(' ', ''))
```

```python
data
```

```python
# renombramos algunos nombres de las columnas
nuevos_nombres={'Table1.HumanDevelopmentIndexanditscomponents':'Country','Unnamed:2':'HDI_Value','Unnamed:4':'Life_expectancy','Unnamed:6':'Expected_years_of_schooling','Unnamed:8':'Mean_years_of_schooling','Unnamed:10':'GNI_per','Unnamed:12':'GNI_rank','Unnamed:14':'HDI_rank'}
data_renombrado=data.rename(columns=nuevos_nombres,inplace=True)
```

```python
data
```

```python
# lista de nombres de columnas a eliminar (datos que no necesitamos)
columnas_a_eliminar = ['Back', 'Unnamed:3','Unnamed:5', 'Unnamed:7', 'Unnamed:9','Unnamed:11', 'Unnamed:13']
data.drop(columns=columnas_a_eliminar, inplace=True)
```
